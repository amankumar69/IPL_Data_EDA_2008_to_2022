```python
plt.figure(figsize=(4,4))  # Create a new plot with a size of 7x7 inches
plt.pie(
    list(batting_decision['WinningTeam'].value_counts()),  # Provide the counts of wins for each team as data
    labels=list(batting_decision['WinningTeam'].value_counts().keys()),  # Assign team names as labels
    autopct='%0.1f%%'  # Display the percentage of each slice with one decimal point
)
plt.show()  # Display the pie chart
```

```python
#batting_second = ipl[(ipl['TossDecision'] == 'bat') & (ipl['WonBy'] == 'Runs')]
batting_second = ipl[(ipl['TossDecision'] == 'field') & (ipl['WonBy'] == 'Wickets')]
```

```python
batting_second.head()
```

```python
batting_second = ipl[(ipl['TossDecision'] == 'field') & (ipl['WonBy'] == 'Wickets')]
# team who bat in second inning and won the matches by Wickets
```

```python
# create a histogram who field first and won the match 
plt.figure(figsize=(5, 7))
batting_second = ipl[(ipl['TossDecision'] == 'field') & (ipl['WonBy'] == 'Wickets')]
plt.hist(batting_second['WinningTeam'], bins=20) # bins use to make shape (more value like 40,50 make size smaller and small value make size larger in chart)
plt.xlabel('Won By Wickets')  # Update x-axis label
plt.ylabel('Frequency')  # Update y-axis label
plt.title('Distribution of Matches Won by Wickets (Fielding First)')  # Update title
plt.grid(True)
plt.show()
```

```python
batting_second['WinningTeam'].value_counts() # teams won the matches who play in second inning or who choose field first
```

```python
# Making a bar plot for distribution of most wins after batting second
plt.figure(figsize=(6,7))    
plt.bar( list(batting_second ['WinningTeam'].value_counts() [0:3].keys()),
        list(batting_second ['WinningTeam'].value_counts() [0:3]),
        color=["red", "yellow", "blue"] 
        )
plt.title('Distribution of Matches Won by Wickets (Fielding First) ')
plt.show()
```

```python
# Making a pie chart for distribution of most wins after batting second
plt.figure(figsize=(7,7))
plt.pie(
    list(batting_second ['WinningTeam'].value_counts()),
    labels= list(batting_second ['WinningTeam'].value_counts().keys()),
    autopct='%0.1f%%'
    )
plt.show()
```
